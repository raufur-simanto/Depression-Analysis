```python
## getting columns with caterogical data
cols_with_object = [col for col in data_reduced.columns if data_reduced[col].dtypes=='object']
print(len(cols_with_object), cols_with_object)
```

```python
## check for null values
data_with_obj = data_reduced[cols_with_object]
#data_with_obj.isnull().sum()
```

## Dealing with categorical data

```python
data_updated = data_reduced.copy()
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
data_updated[cols_with_object] = encoder.fit_transform(data_reduced[cols_with_object])
```

## Looking for correlations

```python
corr_matrix = data_updated.corr()
corr_matrix['How are you feeling right now?'].sort_values(ascending=False)
```

```python
## define target and features
y = data_updated['How are you feeling right now?']
X = data_updated.drop(['How are you feeling right now?', 'Gender', 'Age', 'Relationship status','Which year are you in?','Have you used any social media within the last 6 hours?'], axis=1)
```

## Split data into train and validation

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=1)
```

```python
print(X_train.shape, X_valid.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

def training(ctype,X_train, y_train, X_valid):
    model = DecisionTreeClassifier(criterion = ctype,random_state = 100,max_depth=100, max_leaf_nodes=100)
    model.fit(X_train, y_train)
    y_predict = model.predict(X_valid)
    return y_predict

    
```

```python
def get_accuracy(y_predict, y_valid):
    from sklearn.metrics import accuracy_score
    accuracy = accuracy_score(y_predict, y_valid) * 100
    return accuracy

def get_confusion_matrix(y_predict, y_valid):
    from sklearn.metrics import confusion_matrix
    matrix = confusion_matrix(y_valid, y_predict)
    return matrix
    
```

```python
def get_report(y_predict, y_valid):
    from sklearn.metrics import classification_report
    return classification_report(y_valid, y_predict)
```

```python
## Result using criterion 'entropy'
y_predict = training('entropy', X_train, y_train, X_valid)
print("Prediction:\n",y_predict,'\n')
print('Accuracy:\n',get_accuracy(y_predict, y_valid), '\n')
print('Confussion matrix:\n', get_confusion_matrix(y_predict, y_valid), '\n')
print("Classification report:\n",get_report(y_predict, y_valid), '\n')
```

```python
## Result using criterion 'gini'
y_predict = training('gini', X_train, y_train, X_valid)
print("Prediction:\n",y_predict,'\n')
print('Accuracy:\n',get_accuracy(y_predict, y_valid), '\n')
print('Confussion matrix:\n', get_confusion_matrix(y_predict, y_valid), '\n')
print("Classification report:\n",get_report(y_predict, y_valid), '\n')
```
