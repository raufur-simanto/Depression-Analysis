```python
import pandas as pd
```

```python
data = pd.read_csv('Copy of Depression and Happiness Factor Analysis.csv')
data = data.drop(['Timestamp'], axis=1)
data.head()
```

```python
## delete record containing null value
data = data.dropna(axis=0)
```

```python
data.info()
```

```python
data.columns
```

```python
import numpy as np
transactions = data.to_numpy().tolist()
```

```python
transactions
print(len(transactions))
```

```python
## convert all values to object
for t in transactions:
    for i in range(len(t)):
        s = ''
        if type(t[i]) != str:
            s = str(t[i])
            t[i] = s
```

```python
for t in transactions:
    for item in t:
        if type(item)!= str:
            print(item)
```

## Encoding

```python
from mlxtend.preprocessing import TransactionEncoder
encoder = TransactionEncoder()
encoded_transactions = encoder.fit(transactions).transform(transactions)
```

```python
print(encoded_transactions)
```

```python
df = pd.DataFrame(encoded_transactions, columns=encoder.columns_) # create a dataframe
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

## Get frequent items

```python
from mlxtend.frequent_patterns import apriori
# Using Apriori algo to create frequent items
frequent_items = apriori(df, min_support=0.50, use_colnames=True)
print(frequent_items)
```

## Association Rule Mining

```python
from mlxtend.frequent_patterns import association_rules
# Association Rule Mining
result = association_rules(
frequent_items, metric='confidence', min_threshold=.60)
```

```python
result
```

```python
new_result = result.loc[:, ['antecedents',
'consequents', 'support', 'confidence']]
new_result
```
